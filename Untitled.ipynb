```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_json('cleaned.json')
df
```

```python
df['book_title']=df['book_title'].apply(lambda x: x.strip())
df.head()
```

```python
def recommender(genre):
    """
    recommend top rated book from same genre. 
    """
    return df[df['genre']==genre].sort_values(by='ratings',ascending=False)
```

```python
recommender('Classics')
```

```python
df['genre'].value_counts()
```

```python
recommender('Fiction')
```

```python
def recommender1(genre):
    
    """
    recommend top rated book from same genre. 
    if only single book form that genre ->return randomly from all time top 10 book
    """
    top=df[df['genre']==genre].sort_values(by='ratings',ascending=False)
    if top.shape[0]>1:
        return dict(top.iloc[0])
    else:
        return dict(df.sort_values(by='ratings',ascending=False).iloc[:10].sample(1).squeeze())
        
```

```python
recommender1('Fiction')
```

```python
recommender1('Religion')
```
